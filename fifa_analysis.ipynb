```python
# top goal scorers
df_sorted = df.sort_values('goals_scored_last_4y', ascending=False)
df_sorted[['team', 'goals_scored_last_4y']].head(10)
```

```python
# simple scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['goals_scored_last_4y'], df['goals_received_last_4y'])
plt.xlabel('Goals Scored')
plt.ylabel('Goals Received')
plt.title('Goals Scored vs Received')
plt.show()
```

```python
# top 10 by market value
top_teams = df.nlargest(10, 'squad_total_market_value_eur')

plt.figure(figsize=(10,5))
plt.bar(top_teams['team'], top_teams['squad_total_market_value_eur']/1000000000)
plt.xlabel('Team')
plt.ylabel('Market Value (Billion Euro)')
plt.xticks(rotation=45)
plt.title('Most Expensive Teams')
plt.show()
```

```python
# average goals by continent
continent_avg = df.groupby('continent')['goals_scored_last_4y'].mean()
print(continent_avg)

# simple bar chart
continent_avg.plot(kind='bar')
plt.xlabel('Continent')
plt.ylabel('Average Goals')
plt.title('Which continent scores more?')
plt.show()
```

```python
# do hosts score more?
hosts = df[df['is_host'] == 1]
not_hosts = df[df['is_host'] == 0]

print("Host teams average goals:", hosts['goals_scored_last_4y'].mean())
print("Non-host average goals:", not_hosts['goals_scored_last_4y'].mean())
```

```python
# seeing which numbers relate to each other
numeric_df = df.select_dtypes(include=[np.number])
numeric_df.corr()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# picking some columns to predict goals
X = df[['fifa_rank_pre_tournament', 'squad_total_market_value_eur']]
y = df['goals_scored_last_4y']

# removing empty rows (just in case)
X = X.dropna()
y = y[X.index]

# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# train model
model = LinearRegression()
model.fit(X_train, y_train)

# test it
score = model.score(X_test, y_test)
print("Model accuracy (R squared):", score)
```

```python
# predict and plot
y_pred = model.predict(X_test)

plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Goals')
plt.ylabel('Predicted Goals')
plt.title('Actual vs Predicted Goals')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```
